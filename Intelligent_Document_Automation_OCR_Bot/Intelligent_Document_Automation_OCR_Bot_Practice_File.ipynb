Our Problem Statement

Organizations receive thousands of document images such as receipts, invoices, bills, and forms every day. Manually reading these documents and entering the information into a database is time-consuming, costly, and error-prone. The goal of this project is to automate this process by extracting text from document images using OCR and converting it into structured information using Gemini AI, reducing manual effort and improving efficiency.

OCR = Optical Character Recognition

It converts text inside an image into machine-readable text.
OCR does not understand the meaning. It only recognizes characters.

What is Intelligent Document Processing (IDP)?

OCR says:

"Here is all the text I found."

IDP says:

"I know which text is the invoice number, which is the date, which is the total amount, and which is the vendor."

That's the key difference.

OCR	IDP
Reads text	Understands text-
Plain text output	Structured data-
No context	Uses context-
Character recognition	Information extraction

Where does Gemini fit?

OCR produces:

ABC STORE

TOTAL 560

GST 28

DATE 12/06/2026

Gemini transforms that into:

{
  "store_name": "ABC STORE",
  "date": "12/06/2026",
  "total": 560,
  "gst": 28
}

Gemini adds understanding and structure to the OCR output.

What is cv2.cvtColor()?

This is one of the most frequently used OpenCV functions.
Here:

source_image → the image you want to convert (image)
conversion_type → what conversion you want (cv2.COLOR_BGR2RGB)

So we're saying:

"Take the OpenCV image stored in BGR format and convert it to RGB format."

## PreProcessing

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print(gray.shape)
print(gray.dtype)
```

Q: Why is grayscale used before OCR?

A strong answer:

OCR focuses on recognizing the shapes of characters rather than their colors. Converting an image to grayscale removes unnecessary color information, reduces the image to a single channel, lowers computational complexity, and simplifies preprocessing techniques such as thresholding and noise removal.

##Thresholding

```python
binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

plt.figure(figsize=(8,10))
plt.imshow(binary, cmap="gray")
plt.axis("off")
plt.title("Binary Image (Otsu Thresholding)")
plt.show()
```

What does thresholding do?

Thresholding chooses one threshold value.

Suppose the threshold is 128.

Then OpenCV asks every pixel:

"Are you greater than 128?"

If YES:

→ Make it 255 (White)

If NO:

→ Make it 0 (Black)

```python
binary = cv2.threshold(
    gray,
    0,
    255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)[1]

plt.figure(figsize=(8,10))
plt.imshow(binary, cmap="gray")
plt.axis("off")
plt.title("Binary Image")
plt.show()
```
