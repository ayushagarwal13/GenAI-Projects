## Dataset

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("urbikn/sroie-datasetv2")

print("Path to dataset files:", path)
```

## Explore the dataset structure

```python
import os

# Display all folders and files
for root, dirs, files in os.walk(path):
    print(root)
    for file in files[:5]:   # Show only first 5 files
        print("   ", file)
    print("-" * 50)
```

## Count the images

```python
import os

image_extensions = (".jpg", ".jpeg", ".png")

image_count = 0

for root, dirs, files in os.walk(path):
    for file in files:
        if file.lower().endswith(image_extensions):
            image_count += 1

print("Total Images:", image_count)
```

## Finding One Sample Images

```python
import os

image_path = None

for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".jpg"):
            image_path = os.path.join(root, file)
            break
    if image_path:
        break

print(image_path)
```

Step 1: Convert BGR to RGB

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

## PreProcessing

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print(gray.shape)
print(gray.dtype)
```

##Thresholding

```python
binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

plt.figure(figsize=(8,10))
plt.imshow(binary, cmap="gray")
plt.axis("off")
plt.title("Binary Image (Otsu Thresholding)")
plt.show()
```

```python
binary = cv2.threshold(
    gray,
    0,
    255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)[1]

plt.figure(figsize=(8,10))
plt.imshow(binary, cmap="gray")
plt.axis("off")
plt.title("Binary Image")
plt.show()
```
